## Disyuntiva

Nos encontramos muchas veces que a la hora de comprar un vino, si no lo conocemos previamente, no sabemos si va a ser bueno o malo. A menudo la gente opta por comprar un vino caro, dando por hecho que será un buen vino. Sin embargo, todos conocemos a alguien que dice saber de un vino muy bueno y a la vez barato. Nos encontramos por tanto ante una disyuntiva: ¿Son todos los vinos buenos caros?, ¿Cuanto mejor es un vino, más alto es su precio?

## Objetivos del trabajo

Queremos resolver en este caso la correlación existente entre el precio de un vino y su calidad. El precio es un dato objetivo, pero nos encontramos un problema con la calidad del vino, ya que es un dato subjetivo. Para solucionarlo, vamos a dar por válida la puntuación que otorga un experto en la materia a cada vino

Contamos con los datos de distintos vinos del mundo, entre los que se encuentran su precio, su puntuación por un experto, y otros datos que pueden ser relevantes, como la zona de procedencia, o el tipo de uva con el que se elabora. Los vinos no son todos del mismo año, primero por razones lógicas de maduración (no pasan el mismo tiempo en barrica un crianza que un reserva, por ejemplo) y segundo porque hay vinos de distintas añadas. 

Los datos disponibles son vinos de distintas partes del mundo. Lógicamente hay más registros de los paises con mayor tradición vinícola, como pueden ser Francia, España, Italia o Chile. Aunque en un principio vamos a analizar todos los datos con un análisis exploratorio, para realizar el estudio en profundidad sólo vamos a tomar en consideración los vinos españoles, que son los que personalmente me interesa estudiar.

## Descripción de las variables contenidas en el dataset

El dataset usado proviene de la plataforma usada por muchos data scientists llamada Kaggle. La muestra cuenta con 129.970 registros de vinos. Las variables que encontramos en la muestra son las siguientes:

   **Country**: país de origen del vino
   
   
   **Description**: breve descripción sobre los aromas del vino
   
   
   **Designation**: viñedo del que provienen las uvas que elaboran el vino
   
   
   **Points**: la cantidad de puntos con la que WineEnthusiast calificó al vino en una escala de 1-100
   
   
   **Price**: es el precio de una botella de vino
   
   
   **Province**: es la provincia o estado de procedencia del vino
   
   
   **Region_1**: es el área vitivinícola en una provincia o estado
   
   
   **Region_2**: si existe una región más específica dentro del área vitivinícola
   
   
   **Taster_name**: nombre del catador del vino
   
   
   **Taster_twitter_handle**: cuenta de Twitter asociada al catador del vino
   
   
   **Title**: nombre del vino (incluye la añada)
   
   
   **Variety**: variedad de uva empleada en la elaboracion del vino
   
   
   **Winery**: la bodega que elaboró el vino

## Preparación de los datos para el análisis

Haciendo y visualizando un resumen de los datos, vemos que va a haber problemas a la hora de tratarlos y buscar correlaciones porque la mayoría son datos categóricos.
Además, vemos que alguno de estos datos categóricos están muy poco homogeneizados. Es el caso de la variedad de la uva, que contiene muchas observaciones imprecisas como "Tinto fino" o "Vino de mesa". Además, esta variable está relacionada con la zona geográfica de procedencia del vino, ya que cada zona da su propia variedad de uva. Es por tanto que esta variable va a ser eliminada de la muestra.

```python
datos_spain_df.drop(['variety'], axis='columns', inplace=True)
```

```python
datos_spain_df.head(5)
```

Vemos que otro gran problema al que nos enfrentamos, es que hay un numero muy alto de denominaciones de origen. Como es una variable categórica que habrá que transformar en dummie, nos encontraremos con una presencia muy alta de variables ficticias, lo cual complica mucho el tratamiento de los datos. Es por ello que sólo vamos a tomar en consideranción para el estudio aquellos vinos pertenecientes a alguna de las denominaciones de origen con mayor nombre en el mundo (Rioja, Ribera del Duero, Cava, Rias Baixas, Rueda, Toro, Bierzo, Vinos de la Tierra de Castilla y León, Jerez, La Mancha, Penedes y Valencia)

```python
datos_vinos = datos_spain_df.loc[datos_spain_df['region_1'].isin(['Rioja', 'Ribera del Duero', 'Cava, Rias Baixas', 'Rueda', 'Toro', 'Bierzo', 'Vinos de la Tierra de Castilla y León', 'Jerez', 'La Mancha', 'Penedes', 'Valencia'])]
```

```python
datos_vinos.count()
```

```python
datos_vinos.head(20)
```

### Creación de variables dummies

Para poder tratar las variables categóricas en los siguientes cálculos, las convertimos en dummies:

```python
## Dummies para province, tomando los datos de datos_vinos
dummies_province = pd.get_dummies(datos_vinos['province']).rename(columns=lambda x: 'province_' + str(x))
df_province = pd.concat([datos_vinos, dummies_province], axis=1)
df_province = pd.DataFrame(df_province)
datos_vinos_df = df_province # Volvemos a cambiarle el nombre al dataframe
datos_vinos_df.head(10)
```

```python
## Dummies para region_1, tomando los datos del dataframe que acabamos de modificar
dummies_region_1 = pd.get_dummies(datos_vinos_df['region_1']).rename(columns=lambda x: 'region_1_' + str(x))
df_region_1 = pd.concat([datos_vinos_df, dummies_region_1], axis=1)
df_region_1 = pd.DataFrame(df_region_1)
datos_vinos_df = df_region_1 # Volvemos a cambiarle el nombre al dataframe
datos_vinos_df.head(10)
```

### Matriz de correlaciones

```python
matriz_corr = datos_vinos_df.corr()
plt.figure(figsize=(7,7))
sns.heatmap(round(matriz_corr, 2), annot=True, fmt=".2g", 
            cmap = sns.light_palette("darkblue", as_cmap=True))
```

Podemos observar, como era previsible, que existe una alta correlación entre las variables "region_1" y las variables "province", ya que dentro de cada zona geográfica hay unas denominaciones de origen muy específicas.
Vemos también la existencia de correlación entre el precio y la puntuación del vino, que era una de las cosas que se querían exxplicar en este estudio

## Modelo predictivo

A continuación vamos a realizar un modelo predictivo que trate de explicar la puntuación de un vino en función de otras variables. Comenzaremos limpiando un poco los datos:

### Limpieza de datos

Como hemos visto anteriormente, las variables "region_1" y "province" vienen a explicar lo mismo, con la diferencia de que "region_1" lo hace de forma más detallada. Es por este motivo que procedemos a la eliminación de la variables "province" así como de sus dummies


(como no deja eliminar las columnas, elimino en una tabla anterior la variable "province" y vuelvo a crear las dummies para "region_1")

```python
#Primero elimino la columna "province"
datos_vinos.drop(['province'], axis='columns', inplace=True)
datos_vinos.head(10)
```

```python
# Y después creo de nuevo las dummies para "region_1"
dummies_region_1 = pd.get_dummies(datos_vinos['region_1']).rename(columns=lambda x: 'region_1_' + str(x))
df_region_1 = pd.concat([datos_vinos, dummies_region_1], axis=1)
df_region_1 = pd.DataFrame(df_region_1)
datos_vinos_prediccion = df_region_1 # Volvemos a cambiarle el nombre al dataframe
datos_vinos_prediccion.head(10)
```

A continuación nos damos cuenta de que para hacer Lasso en el futuro, vamos a tener problemas con las variables categóricas ("title" y "region_1"), por lo que vamos a crear un nuevo dataframe (datos_para_lasso) que excluya estas variables

```python
#Creamos los dos data frame separados, cada uno con variables únicamente numéricas 
#Hay que hacerlo así porque da error al eliminar algunas columnas
h = datos_vinos_prediccion.iloc[:, 4:12] #Variables dummie de region_1
i = datos_vinos_prediccion.iloc[:, 0:2] #Variables que la explican (las demás)

#Juntamos los dataframe en uno solo, al que llamamos datos_para_lasso
datos_para_lasso = pd.concat([i, h], axis=1)
```

### Visualización de datos

Volvemos a ver la matriz de correlaciones:

```python
matriz_corr = datos_para_lasso.corr()
plt.figure(figsize=(7,7))
sns.heatmap(round(matriz_corr, 2), annot=True, fmt=".2g", 
            cmap = sns.light_palette("darkblue", as_cmap=True))
```

Y ahora graficamos la reación entre el precio y la puntuación del vino

```python
plt.scatter(datos_para_lasso['price'],datos_para_lasso['points'])
plt.ylabel('Puntuación')
plt.xlabel('Precio')
```

### Datos training y datos test

Separamos los datos en training y test para realizar un modelo y poder probarlo después

```python
# Preparamos la librería
from sklearn.model_selection import train_test_split

#Creamos la división

y = datos_para_lasso.iloc[:, 0] #Variable que queremos explicar ("points")
x = datos_para_lasso.loc[:, datos_para_lasso.columns != "points"] #Variables que la explican (las demás)

#Hacemos una partición del 70% para prueba y 30% para test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=123)

# Vemos las dimensiones de la partición
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

### Lasso

Procedemos a hacer un análisis de regresión con el método Lasso

```python
## Cargamos las librerías

from sklearn import linear_model
from sklearn.linear_model import LassoCV
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

## Formulamos el modelo

lassocv = LassoCV(alphas = None, cv = 10, max_iter = 100000, normalize = True)
lassocv.fit(x_train, y_train)
lassocv.alpha_

# Realizamos la regresion

model_lasso = Lasso(alpha=lassocv.alpha_) # funcion lasso
model_lasso.fit(x_train, y_train) # ajuste sobre el train
pred_lasso = model_lasso.predict(x_test) # prediccion sobre el test
print(np.sqrt(mean_squared_error(y_test,pred_lasso))) # error sobre el test
error_lasso = np.sqrt(mean_squared_error(y_test,pred_lasso))
print(r2_score(y_test, pred_lasso)) # R^2 sobre el test
```

Vemos que tenemos un error de 2.85 puntos a la hora de determinar la calidad del vino.

## Conclusion

Vemos que existe una correlación positiva entre la calidad de un vino y su precio. Esta correlación se ve influenciada, aunque sea levemente, con la denominación de origen del vino. Podemos determinar por tanto que, conociendo el precio del vino, y su denominación de origen, podemos predecir con bastante exactitud su calidad.
